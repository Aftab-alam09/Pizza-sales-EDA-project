##                                                               Pizza Sales EDA
### Here, The data has been analysed with only Python, pandas and matplotlib.while in this EDA the only quantity aspect has been considered is quantity not the price of pizza.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file=pd.read_csv('pizza_sales.csv')
df=pd.DataFrame(file)
```

```python
df.head()
```

```python
### finding the null value in the data frame.
df.isnull().sum()
```

```python
### Finding out the unique value of each column.
df['quantity'].unique()
```

```python
df['pizza_category'].unique()
```

```python
### Here the pizza ingredients aren't neccessary and doesn't contribute the value in the data set. So, it is dropped
df.drop('pizza_ingredients',axis=1,inplace=True)
```

```python
df['pizza_name'].unique()
```

```python
df['unit_price'].unique()
```

```python
## In this data price aren't being considered then needs to drop.
df['total_price'].unique()
df.drop('total_price',axis=1,inplace=True)
```

```python
df.head()
```

## KEY REQUIREMENT FOR EDA
Quantity with Category.

Most sold Pizza Size.

Most Sold pizza name.

```python
Total_Q_Sales=df['quantity'].count()
```

```python
Total_salesby_cate=df.groupby('pizza_category')['quantity'].sum()
```

```python
plt.xlabel('Pizza Category')
plt.ylabel('Total Sales')
plt.title('Total sales by category')
plt.bar(Total_salesby_cate.index,Total_salesby_cate.values,color='orange')
plt.figure(figsize=(10,8))
```

## Observation 
The Most sold pizza is classic having quantity over 14000 units.

```python
Total_salesby_size=df.groupby('pizza_size')['quantity'].sum()
Total_salesby_size.head(4)
```

```python
plt.title('Sales by Pizza Size')
plt.xlabel('Pizza Size')
plt.ylabel('Quantity of Pizza')
plt.barh(Total_salesby_size.index,Total_salesby_size.values)
```

```python
## Observation
The L size pizza are the most sold pizza.
```

```python
plt.title('Sales by Pizza Size')
plt.pie(Total_salesby_size.values,labels=Total_salesby_size.index,labeldistance=1.1,
        startangle=90,shadow=0.2,explode=[0.1,0,0,0,0],colors=['lightblue','blue','yellow','red','gray'],autopct='%1.1f%%')
```

## Observation 
The L size contributes 38.2% in the sales, M with 31.5%, S with 29.1% and XL with 0.1%

## KEY REQUIREMENT FOR EDA
Customers by pizza quantity.

```python
plt.title('Most sold pizza Quantity')
plt.xlabel('Pizza pieces')
plt.ylabel('Total Number of Quantity by pieces in a year')
plt.scatter(df['quantity'],df['order_id'],color='blue',label='pizza_sales')
plt.grid(True)
```

```python
## Observation
Most of the Customers are buying 1 or 2 pices of items/pizza.
```

```python
df.head()
```

```python
df['months']=df['order_date'].str.split('-').str[1]
```

```python
df.head()
```

```python
## sales by months
sales_by_months=df.groupby('months')['quantity'].sum()
sales_by_months
```

```python
## Bins color with different colors and then observation.
```

```python
plt.title('MONTHLY SALES')
plt.xlabel('Months')
plt.ylabel('Quantity')
plt.bar(sales_by_months.index,sales_by_months.values,color=['orange','yellow','cyan','lightpink','orange','yellow','cyan','lightpink','orange','yellow','cyan','lightpink'])
plt.grid(False)
```

## Observation
There is almost no spike in one month and sckewed in another. The sales are almost same in each months
